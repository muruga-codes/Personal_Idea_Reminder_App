What is a List?

A list stores multiple values.

```python
a=["cricket","ODI","T20","IPL","CPL"]
print(a)
```

Access List Elements

```python
a=["cricket","ODI","T20","IPL","CPL"]
print(a[0])
print(a[3])
```

Add New Element

```python
b=[]
b.append("Testmatch")
b.append("WPL")
print(b)
```

User Adds Elements

```python
b=[]
a=input("enter a:")
b.append(a)
print(b)
```

Add Multiple Elements

```python
b=[]
for i in range(3):
    a=input("enter a:")
    b.append(a)
print("\n new element")
print(b)
```

Display Elements Nicely

```python
b=["BMW","BMW M4","Rolls"]
for i in b:
    print(i)
```

Number the Elements

```python
b=["BMW","BMW M4","Rolls"]
for item,a in enumerate(b,start=1):
    print(item,".",a)
```

Without enumerate()

```python
b=["BMW","BMW M4","Rolls"]
index=1
for i in b:
    print(index,".",i)
    index+=1
```

Search an Element

```python
list=["BMW","BMW M4","Rolls","Roys"]
search=input("enter the element to search:")
if search in list:
    print("search is found")
else:
    print("search not found")
```

Delete an Element

```python
list=["BMW","BMW M4","Rolls","Roys"]
input_1=input("enter the element to delete:")
if input_1 in list:
    list.remove(input_1)
    print("Element is delete")
else:
    print("element is not delete")
print(list)
```
