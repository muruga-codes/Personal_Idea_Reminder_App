Task 6

Create a menu

```python
def menu():
    while True:
        Choice=input("enter your choice:")
        if Choice=="1":
            print("Add idea")
            break
        elif Choice=="2":
            print("Search idea")
            break
        elif Choice=="3":
            print("Delete idea")
            break
        else:
            print("exit")
menu()
```
